# IMPORT DAS BIBLIOTECAS

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import scipy.stats as sct
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import missingno as msno
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import warnings
warnings.filterwarnings('ignore')
```

# LOAD  DOS DATASETS TRAIN E TEST

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

# IGUALANDO AS VARIÁVEIS NO TRAIN E NO TEST

```python
df_train.shape,df_test.shape
```

```python
cols_test = df_test.columns.sort_values()
cols_test = cols_test.append(pd.Index(['IN_TREINEIRO'])).sort_values()
df_train = df_train[cols_test]
df_train.shape,df_test.shape
```

# LIMPEZA DOS DADOS

## BINARIZANDO A VARIÁVEL TP_SEXO

```python
df_train.TP_SEXO = df_train['TP_SEXO'].map({'M':0,'F':1})
df_test.TP_SEXO = df_test['TP_SEXO'].map({'M':0,'F':1})
```

## ANÁLISE DOS VALORE NULOS (NaN) 

```python
df_train.info()
print()
df_test.info()
```

### Dropando colunas com qtd significativa de NaN:

```python
df_train.drop(columns=['TP_DEPENDENCIA_ADM_ESC','TP_ENSINO'],axis=1,inplace=True)
df_test.drop(columns=['TP_DEPENDENCIA_ADM_ESC','TP_ENSINO'],axis=1,inplace=True)
df_train.shape,df_test.shape
```

### Substituindo NaN por 0 nas variaveis relacionadas às Provas:

As provas de LC e MT foram realizadas no mesmo dia (06/11/2016), e as provas de CN e CH no dia 05/11/16
Todas as variaveis relacionadas as Provas e Redacao possuem numeros de NaN identicos entre LC e MT, e CN e CH, inclusive as variaveis TP_PRESENCA. Desta forma podemos concluir que os alunos não estiveram presentes  ou no primeiro dia, ou no segundo ou em ambos. Logo podemos substituir os NaN por 0.

```python
df_train.fillna(value=0,axis=1,inplace=True)
df_test.fillna(value=0,axis=1,inplace=True)
```

```python
msno.matrix(df_train),msno.matrix(df_test)
```

## CORREÇÃO DOS TIPOS DAS VARIÁVEIS

```python
df_train.dtypes
```

```python
cols_object_train = df_train.select_dtypes(exclude='float64').columns
df_train[cols_object_train] = df_train[cols_object_train].astype('object')
df_train.TP_STATUS_REDACAO = df_train['TP_STATUS_REDACAO'].astype('object')
df_train.NU_IDADE = df_train['NU_IDADE'].astype('int64')
df_train.IN_TREINEIRO = df_train['IN_TREINEIRO'].astype('int64')
df_train.dtypes
```

```python
cols_object_test = df_test.select_dtypes(exclude='float64').columns
df_test[cols_object_test] = df_test[cols_object_test].astype('object')
df_test.TP_STATUS_REDACAO = df_test['TP_STATUS_REDACAO'].astype('object')
df_test.NU_IDADE = df_test['NU_IDADE'].astype('int64')
df_test.dtypes
```

# PREPARAÇÃO DOS DADOS PARA CLASSIFICAÇÃO

## DUMMIZAÇÃO DAS VARIÁVEIS CATEGÓRICAS

```python
df_to_dummie = pd.concat([df_train,df_test],axis=0)
df_to_dummie=df_to_dummie.drop(['NU_INSCRICAO'],axis=1)
df_to_dummie=pd.get_dummies(df_to_dummie,drop_first=True)
df_to_dummie.head()
```

```python
df_train_dummie = df_to_dummie[pd.notna(df_to_dummie['IN_TREINEIRO'])]
df_test_dummie = df_to_dummie[pd.isnull(df_to_dummie['IN_TREINEIRO'])]
df_test_dummie = df_test_dummie.drop(['IN_TREINEIRO'],axis=1)

print(df_train_dummie.shape,df_train.shape[0])
print(df_test_dummie.shape,df_test.shape[0])
```

## BALANCEAMENTO DA VARIÁVEL TARGET : IN_TREINEIRO

```python
sm = SMOTE(sampling_strategy='minority', random_state=42)
X, y = sm.fit_resample(df_train_dummie.drop('IN_TREINEIRO',axis=1),df_train_dummie['IN_TREINEIRO'])
```

```python
sns.countplot(y)
```

# COMBINAÇÃO DE MODELOS CLASSIFICADORES

Serão testados 6 modelos de classificação, sendo que num primeiro momento as predições serão definidas através da combinação dos resultados de todos eles, em um conceito de votação simples (Ensembles). Posteriormente será utilizado o resultado do modelo com o melhor score, para efeito de comparação da melhor acurácia após submissão na Codenation.

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state=1)
score_list = []
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## LOGISTIC REGRESSION 

```python
lr = LogisticRegression()

lr.fit(x_train,y_train)

print('Logistic Regression Score : ',lr.score(x_test,y_test))#Ratio of correct predictions

score_list.append(lr.score(x_test,y_test))
```

```python
#this is our real prediction part

lr.fit(X,y)

lr_prediction = lr.predict(df_test_dummie)
```

## K-NEAREST NEIGHBORS 

```python
knn = KNeighborsClassifier(n_neighbors = 3)

knn.fit(x_train,y_train)

print('K-Nearest Neighbors Score : ',knn.score(x_test,y_test))

score_list.append(knn.score(x_test,y_test))
```

```python
knn.fit(X,y)

knn_prediction = knn.predict(df_test_dummie)
```

## SUPPORT VECTOR MACHINE

```python
svm = SVC(random_state = 1)

svm.fit(x_train,y_train)

print('Super Vector Machine Score : ',svm.score(x_test,y_test))

score_list.append(svm.score(x_test,y_test))
```

```python
svm.fit(X,y)

svm_prediction = svm.predict(df_test_dummie)
```

## NAIVE BAYES

```python
nb = GaussianNB()

nb.fit(x_train,y_train)

print('Naive Bayes Score : ',nb.score(x_test,y_test))

score_list.append(nb.score(x_test,y_test))
```

```python
nb.fit(X,y)

nb_prediction = nb.predict(df_test_dummie)
```

## DECISION TREE

```python
dt = DecisionTreeClassifier()

dt.fit(x_train,y_train)

print('Decision Tree Score : ',dt.score(x_test,y_test))

score_list.append(dt.score(x_test,y_test))
```

```python
dt.fit(X,y)

dt_prediction = dt.predict(df_test_dummie)
```

## RANDOM FOREST

```python
rf = RandomForestClassifier(n_estimators = 22,random_state = 40)

rf.fit(x_train,y_train)

print('Random Forest Score : ',rf.score(x_test,y_test))

score_list.append(rf.score(x_test,y_test))
```

```python
rf.fit(X,y)

rf_prediction = rf.predict(df_test_dummie)
```

## ENSEMBLE - VOTAÇÃO SIMPLES ENTRE OS MODELOS
PARA DETERMINAR A MELHOR PREDIÇÃO, SERÁ REALIZADA UMA COMPARAÇÀO ENTRE OS MODELOS E A PREDIÇÃO FINAL SERÁ ESCOLHIDA PELA MAIOR VOTAÇÃO ENTRE ELES.

```python
pr_dict = {'Logistic Regression' : lr_prediction,
           'KNN' : knn_prediction,
           'SVM' : svm_prediction,
           'Naive Bayes' : nb_prediction,
           'Decision Tree' : dt_prediction, 
           'Random Forest' : rf_prediction}

all_predictions = pd.DataFrame(pr_dict)

all_predictions
```

```python
final_prediction = [] #final prediction list

#i : range rows , j : range cols

for i in all_predictions.values:
    sum_zero_score = 0 #summary of zero scores
    sum_one_score = 0 #summary of one scores
    
    for j in range(5):
        if i[j]==0:
            sum_zero_score += score_list[j]
        else:
            sum_one_score += score_list[j]
    
    if sum_zero_score >= sum_one_score:
        final_prediction.append(0)
    else:
        final_prediction.append(1)
    
```

## PREPAÇÃO DO ARQUIVO ANSWER.CSV - BASEADO NO PREDICT DO ENSEMBLE

```python
sns.countplot(final_prediction)
```

```python
answer_ensemble = pd.DataFrame({'NU_INSCRICAO':df_test['NU_INSCRICAO'],
          'IN_TREINEIRO':final_prediction})
answer_ensemble.to_csv('answer.csv',index=False)
answer_ensemble.head()
```

**Esta primeira submissão já superou a meta, atingindo o score = 98.030635**

Como a predição acima foi definida a partir do score do Ensemble dos seis modelos testados, na sequência será feita a submissão do predict do modelo que apresentou a melhor acurácia.

# RANDOM FOREST: MODELO QUE APRESENTOU A MELHOR ACURÁCIA

```python
sns.countplot(rf_prediction)
```

```python
answer_rf = pd.DataFrame({'NU_INSCRICAO':df_test['NU_INSCRICAO'],
          'IN_TREINEIRO':rf_prediction})
answer_rf.to_csv('answer.csv',index=False)
answer_rf.head()
```

**O Score obtido utilizando o predict apenas do Random Forest foi igual a 99.759300, superando a sumbissão anterior que utilizou o resultado da votação do Ensemble.**

# CONCLUSÕES

A modelagem deste algoritmo de classificação foi pontuada por duas importantes etapas: **Preparação dos Dados** (tópicos 5 a 8 e subitens) e **Modelagem** (tópicos 9, 10 e subitens).

Ficou clara a importância de uma adequada preparação dos dados para a obtenção de um modelo com elevado nível de assertividade.

Na **Preparação dos Dados** foi realizada correção dos tipos das variáveis, binarização e dummização das variáveis categóricas, análise e tratamento dos valores nulos, e destaque especial para o balanceamento da variável alvo, que foi fundamental para o sucesso do modelo.

Na parte da **Modelagem** foi utilizada a técnica de combinação de classificadores (Ensemble), onde o resultado foi obtido através de um sistema de votação simples (sem pesos) entre 6 modelos classificadores, e o score obtido após a submissão foi bem elevado: **98.030635**.

Mas como a performance entre os modelos oscilou significativamente na fase dos testes, foi realizada uma nova submissão utilizando somente as predições do modelo que apresentou o melhor score nos testes, o Random Forest, e surpreendentemente o score aumentou para **99.759300**.
